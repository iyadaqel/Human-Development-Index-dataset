```python
#This code scraps the website of UN to get all the kpis of the Human Development Index. To run it you'll have to download ChromeDriver from here: 
# https://sites.google.com/a/chromium.org/chromedriver/
# Then reference the .exe/.dmg in the code 

import os
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
input_path = "./input_data/HDI"
os.makedirs(input_path, exist_ok=True)

#The csv file has the url for all the indices. 
HDI = pd.read_csv("HDI_kpis.csv" , header=0)
HDI.head()
```

```python
def getData(obj):
    print("Getting data for " + str(obj['kpi_name']))
    index_dict ={}
    page_url =obj['url']
    options = webdriver.ChromeOptions()
    options.add_argument('headless')
    ####### Change with the direcory  of chrome driver. 
    driver = webdriver.Chrome(executable_path = "C:/Users/IyadAlAquel/Downloads/chromedriver_win32/chromedriver"  , chrome_options=options)
    #######
    a = driver.get(page_url)
    element = WebDriverWait(driver, 60).until(
            EC.presence_of_element_located((By.ID, "table"))
        )
    i=0
    rows = element.find_elements(By.TAG_NAME, "tr") # get all of the rows in the table
    for row in rows:
        header_val = []
        header = row.find_elements(By.TAG_NAME , "th")
        for val in header:
            header_val.append(val.text)
        index_dict[i] = header_val
        if(i>1):
            row_val = []
            col = row.find_elements(By.TAG_NAME, "td") #note: index start from 0, 1 is col 2
            for val in col:
                row_val.append(val.text)
            index_dict[i] = row_val
        i+=1
    df=pd.DataFrame.from_dict(index_dict,orient='index')
    df.columns= df.iloc[1]
    df = df.iloc[2:]
    df.to_csv(input_path + "/" + str(obj['kpi_name']) + ".csv")
    driver.close()
    print("Saved " + str(obj['kpi_name']))
    return ""



HDI.apply(getData , axis=1)
```
